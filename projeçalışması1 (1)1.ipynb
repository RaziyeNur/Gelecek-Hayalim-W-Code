```python
import pandas as pd

# Ondalıklı veriyi tamsayıya (int) dönüştürmek
df['Yıllık Maaş (yaklaşık $USD)'] = df['Yıllık Maaş (yaklaşık $USD)'].astype('Int64')
```

```python
# Eksik verileri 0 ile doldurup Int64'e çevirme
df['Yıllık Maaş (yaklaşık $USD)'] = df['Yıllık Maaş (yaklaşık $USD)'].fillna(0).astype('Int64')
```

```python
df["Yıllık Maaş (yaklaşık $USD)"].fillna(df["Yıllık Maaş (yaklaşık $USD)"].mean(),inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Bağımsız ve bağımlı değişkenleri seçme
X = df[['Yıllık Maaş (yaklaşık $USD)', 'İş Yerinde Makine Öğrenimi Kullanımı']]  # Bağımsız değişkenler
y = df['Cinsiyet']  # Bağımlı değişken

# Eğitim ve test verilerini ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=153)

# Model oluşturma ve eğitme
model = LinearRegression()
model.fit(X_train, y_train)

# Tahmin yapma
lr_y_pred = model.predict(X_test)

# Tahminleri ve gerçek değerleri karşılaştırma
for i in range(0, 10):
    print("Tahmin: ", int(lr_y_pred[i]), "Gerçek: ", y_test.iloc[i])

print("------------------------------")

for i in range(0, len(lr_y_pred)):
    if int(lr_y_pred[i]) == y_test.iloc[i]:
        print("Tahmin: ", int(lr_y_pred[i]), "Gerçek: ", y_test.iloc[i])

print("if bitti")
```

```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import matplotlib.pyplot as plt

# Veri hazırlama
X = df[['Yıllık Maaş (yaklaşık $USD)']].values
y = df['İş Yerinde Makine Öğrenimi Kullanımı'].values

# Polinom regresyon (overfitting için yüksek derece kullanılır)
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, y)

# Tahmin yapma
y_pred = model.predict(X_poly)

# Sonuçları görselleştirme
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score

# Karar ağacı regresyon modeli
model = DecisionTreeRegressor()

# K-Fold çapraz doğrulama
scores = cross_val_score(model, X, y, cv=5)

# Sonuçları yazdırma
print('Doğrulama skorları:', scores)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# Modeli oluşturma ve eğitme
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Tahmin yapma
dt_y_pred = model.predict(X_test)

# Performans metrikleri
accuracy = accuracy_score(y_test, dt_y_pred)
precision = precision_score(y_test, dt_y_pred, average="weighted")
recall = recall_score(y_test, dt_y_pred, average="weighted")
f1 = f1_score(y_test, dt_y_pred, average="weighted")

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# MAE, MSE ve R2 skorlarını hesaplama
mae = mean_absolute_error(lr_y_test, lr_y_pred)
mse = mean_squared_error(lr_y_test, lr_y_pred)
r2 = r2_score(lr_y_test, lr_y_pred)

print(f'MAE: {mae}, MSE: {mse}, R2: {r2}')
```
